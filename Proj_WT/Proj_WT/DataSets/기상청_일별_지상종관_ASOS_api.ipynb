반복문 안에 이 부분은 년별로 불러올 때, 처음 초기화 작업에만 사용하고 runtime 초기화를 한 게 아니면 주석처리해서 계속 쌓을 수 있음

년별로 요청을 하는 경우, 지점별 요청할 url로 코드를 작성했기 때문에 공공데이터에서 보편적으로 요청 페이지 maximum이 1000임을 떠올릴 때 2개년 정도는 반복문으로 돌릴 수 있음

* 예시) 2015~2016년을 불러올 경우

````python
for of, br in zip(office, branch):
    print(f'{StDt[:4]}년도 {of} 데이터 수집을 시작합니다.')
    locals()[f'{of}_df'] = pd.DataFrame()

    for name, code in br.items():
        try:
            url = f"http://apis.data.go.kr/1360000/{SERVICE}/{STYPE}?serviceKey={KEY}&pageNo={PG_NO}&numOfRows={NoRows}&dataType={Dtype[1]}&dataCd={DCd}&dateCd={DtCd}&startDt={StDt}&endDt={EdDt}&stnIds={code}"
            print(f'\t지점 {name} 데이터 수집을 시작합니다.')
            rep = requests.get(url)
            time.sleep(random.randint(2,10))
            origin = rep.json()

            if origin['response']['body']['totalCount']=="0":
                print(f"They don't have Data at {name}")

            else:
                NoRows = origin['response']['body']['totalCount']
                url = f"http://apis.data.go.kr/1360000/{SERVICE}/{STYPE}?serviceKey={KEY}&pageNo={PG_NO}&numOfRows={NoRows}&dataType={Dtype[1]}&dataCd={DCd}&dateCd={DtCd}&startDt={StDt}&endDt={EdDt}&stnIds={code}"
                rep = requests.get(url)
                time.sleep(random.randint(2,15))
                origin = rep.json()
                locals()[f'{of}_df'] = pd.concat([locals()[f'{of}_df'], pd.json_normalize(origin['response']['body']['items']['item'])])

                print(f'\t지점 {name} 데이터 수집이 완료되었습니다.')

        except:
            pass
    print(f'{EdDt[:4]}년도 {of} 데이터 수집이 완료되었습니다')
````
