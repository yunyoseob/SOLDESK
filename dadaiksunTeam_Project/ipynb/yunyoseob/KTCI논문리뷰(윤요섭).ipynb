한국형 관광기후지수(KTCI)의 개발에 관한 연구
--

서론
--

1) TCI 소개

기후는 관광활동에 영향을 주며, 이러한 사실을 우리는 경험적으로 알고 있다.

주로, 기후와 관광간의 관계를 이론적인 측면에서의 논의와 관광자의 활동에 적합한 최적기후에 관한 연구가 있는데, 이러한 최적기후조건을 활용하여,
 관광자를 둘러싼 기후환경을 쉽게 평가하고 목적에 맞게 사용하기 위해 관광기후지수(TCI)를 산출하는 연구가 진행되고 있다.

관광기후지수는 기후환경을 평가하여 관광자가 활동하기에 좋은 시기와 공간적 분포를 제시하며 덩욱이 미래기후시나리오와 결합하여 미래 관광환경의 변화를 예측하는데, 유용한 자료이다.


2) TCI 한계점

관광기후지수(TCI) 는 단지 한 가지의 최적기후조건(기후조건에 대한 선호)을 토대로 대륙별, 지역별 관광자의 이동 흐름을 평가하고 있다는 한계점이 있다.

기후조건에 대한 선호는 지역의 지리나 문화에 따라 달라질 수 있다. 즉, 우리나라 관광자의 경우 야외관광활동을 하기에 좋은 기후조건은 유럽과 다를 수 있다.

더구나 기존 연구들에서 적용하고 있는 열적쾌적성은 서양인을 토대로 산출된 것이며, 가중치와 기후요소의 점수구간은 Mieczkowski(1985)의 주관적인 판단을 토대로 구성되었으므로, 국내에 그대로 적용하기에는 적합하지 않다.

한국의 경우에는 계절마다 다른 기후특성을 가지고 있어서, 계절을 반영한 관광기후지수 개발이 요구되며, 여가활동에 따라 기후요소별 차이가 존재하는지에 대한 검토가 요구 된다.

문헌 고찰
--

1) De Freitas(2003)

관광자의 활동에 적합한 날씨, 기후 또는 최적조건을 열적 측면, 물리적 측면, 미적 측면으로 구분하였다.

1) 열적 측면: 관광자의 열적 쾌적함(편안하게 느끼는)과 기온, 바람, 습도 등이 복합적으로 결합된 것을 의미

2) 물리적 측면: 바람이나 강수처럼 관광활동에 직접적으로 제약요인으로 작용하는 것을 의미

3) 미적 측면: 관광자의 감정이나 경험의 질에 영향을 주는 풍경과 연관된 것으로서
파란하늘, 구름의 양, 햇빛, 시야, 일조시간 등을 의미

2) Gomez-martin(2005)

관광자원으로서 날씨, 기후의 특성을 논의

날씨, 기후의 특성

1) 공유재

2) 운반하거나 저장할 수 없다

3) 공간과 시간에 따라 다양하다.

4) 다양성 측면에서 극한 기후(태풍, 폭우, 홍수 등)은 관광지의 매우 큰 위협요소이다.

5) 날씨, 기후의 요소들은 서로 복합적, 직접적으로 자연과 인간 시스템에
영향을 미치며, 

관광자에게는 날씨, 기후의 물리적(비, 바람 등), 

생리적(기온, 습도 등), 심리적(구름의 양, 일조시간 등) 요소가 영향을 준다.

6) 날씨, 기후는 소비되는 것이 아니기 때문에 재생이 가능하다.

의의: De Freitas(2003)의 논의를 확장하여 지리학 관점에서 논의하였음.

3) Scott & Lemieux(2010)

앞서 두 연구를 체꼐적으로 정리하고 기후의 최적조건 및 과거데이터, 미래 예측 등 다양한 정보의 활용에 대해 논의. 즉, 날씨, 기후가 관광자의 경험의 질에 영향을 주기 때문에, 관광지, 관광활동별로 다양하고, 정확한 날씨, 기후 정보를 관광자에게 
제공할 필요가 있다고 보았음.

의의: 날씨, 기후는 광광자의 활동에 영향을 미치고 만족도를 높이기 위해서 이를 관리할 필요가 있다는 견해를 확장한 것으로 볼 수 있음.

관광자의 최적날씨, 기후 조건
--

관광자의 활동에 적합한 최적날씨, 기후 조건에 대한 연구는 크게 

1. 전문가 기반의 최적기후(TCI 포함) 활용

2. 기후와 관광자 데이터를 활용한 계량적 분석

3. 관광자의 최적기후 인식(설문 조사) 등 세 가지로 구분 할 수 있음

출처: (Scott etal., 2008)



1) 전문가 기반의 접근은 관광활동과 관련된 기후요소를 통합하여 날씨 종류, 기후지수로 나타내고 있다. 그러나 전문가의 주관적인 판단이 포함되고, 관광자에 의해 검증이 되지 않은 한계가 있다.

출처: (De Freitas, 2003; Gomez-Martin. 2006; Scott et al, 2004)

2) 기후와 관광자 데이터를 활용한 계량적 분석은 객관적인 데이터를 활용하여 최적기후를 도출함으로써 연구자의 주관을 배제한다. 하지만, 관광자, 기후에 대해 획득 가능한 데이터가 대부분 국가 단위, 연간, 월 단위이기 때문에 미세한 분석에 한계가 있다.

출처: (Scott et al., 2008)

3) 최적날씨, 기후에 대해 관광자에게 직접 의견을 물어 설문하는 방식은 관광자가 현지에서 지각하는 최적날씨, 기후를 설문하는 형태로 전문가 판단, 데이터 분석 형태보다는 미세하고 현실적인 결과를 도출 할 수 있는 장점이 있다. 그러나 이러한 형태의 연구방법은 모집단 전체에 대한 조사가 불가능하기 때문에 일반화가 어려운 한계점을 안고 있다. 

출처: (Scott et al., 2012)

관광자의 최적날씨, 기후 조건은 관광기후지수 구성의 토대가 된다. 관광자가 활동하기에 가장 좋은 날씨, 기후 조건에 가장 높은 점수를 부여하고 다음 조건에 더 낮은 점수를 순차적으로 부여한다. 이렇게 등급화된 데이터를 지역의 기후조건에 적용하여 공간별, 시기별로 특성을 나타낸다. 더 나아가 지역의 미래 기후변화 시나리오를 토대로 관광환경의 변화를 예측한다.

관광기후지수(TCI)
--

1) TCI

기후는 관광활동의 만족에 영향을 주고 관광자의 관광수요를 발생시킨다.
정책가, 여행업자, 계획가 등이 관광기후환경을 쉽게 평가하고 목적에 맞게 사용하기 위해 지수(index) 형태로 측정되고 있다.

관광자를 둘러싼 기후환경을 평가하기에 처음에는 단일 지수(열지수, 체감기온 등등)이 사용되었으나, 단일지수는 한계점이 있다.

이로 인해 관광기후지수는 지난 30여 년간 다양하게 접근되고 있으며, 그 중  Mieczkowski(1985)가 개발한 관광기후지수(TCI)는 수많은 연구에서 인용되고 있다.

TCI는 도시나, 야외에서 걷기, 관람, 쇼핑 등의 일반관광과 관련된 지수이며, 해변이나 스키 등의 활동에는 적용하지 못하는 한계가 있다.

2) TCI 공식과 변수 설명

TCI = 2(4Cd + Ca +2P + W + 2S)


Cd: 열적쾌적성 (한낮) 

- 월별변수: 최고기온, 최저상대 습도

Ca: 열적 쾌적성 (일평균)

- 월별변수: 평균기온, 평균상대습도

P: 강수(mm)

- 월별변수: 총강수량

W: 풍속(m/s)

- 월별변수: 평균풍속

S: 일조시간(h)

- 월변변수: 일조시간

쾌적지수는 기온과 상대습도를 구간별로 나누어 점수화

TCI의 범위는 -30에서 100까지 10개로 구분된다.

3) TCI 적용 한계

1. 우리나라의 경우 여름철(7,8월)의 경우 최대 휴가기간이지만 TCI 지수가 낮게 나온다. 
(여름철 고온습윤한 기후적 특성과 강수량의 하계 집중적인 영향 때문)

2. 기후변수의 점수화와 가중칙가 관광지나 관광시장에 의해 실증적으로 검증이 되지 않고 연구자의 주관적인 판단에 의해 결정됨. 출처: (Amelung & Viner, 2006; Amelung et al., 2007; Hein 
et al., 2009; Moreno & Amelung, 2009; Perch-Nielsen et al., 2010; 
Scott & McBoyle, 2001; Scott et al., 2004; Whittlesea & Amelung, 
2010)

3.  TCI는 걷기, 쇼핑, 관람 등의 가벼운 관광활동을 하기에 알맞은 기후를 평가하는 것이기 때문에 여름 해변관광이나, 겨울 스포츠와 같은 특정 관광형태에 관한 기후를 평가하기 어렵다.

4. TCI는 관광지의 관광자 수와의 관계를 예측할 수 없다. 출처: (Amelung et al., 2007)


4) TCI 한계 극복 연구

1. TCI의 열적쾌적성 변수의 과대추정된 단점을 극복하기 위해 효율적 기온(ET)를 고안한 열지수로 변경하여 산출 출처:Steadman(1984)

2. 평균이슬점기온으로 부터 계산된 평균수증기압을 활용하고, 한낮의 수증기압을 제외(한낮에는 기온이 증가하면서 상대 습도가 감소하기 때문에 왜곡할 수 있음)하여 열적 쾌적성을 구성 
출처:(PerchNielsen et al., 2010)

3. 설문조사를 통해 해변관광지수 산출

4. Siple & 
Passel(1945)의 체감기온(wind chill index)을 Osczevski & Bluestein 
(2005)의 체감기온(wind chill equivalent temperature)으로 변경하여 TCI
를 구성(Perch-Nielsen et al., 2010).

5. 기존 TCI는 
1951∼1980년의 기후데이터를 활용하고 있으나 최근의 관광기후 환경을 분석하
기 위해 1961∼1990년간의 기후 데이터를 활용

6. 일변 기후 데이터 활용

7. 하나의 기후 모델이 아닌 다섯 가지 기후모델로 결과를 비교하여 기존 단일 기후모델 결과의 한계를 극복


KTCI (한국형 TCI)
--

1. 조사 방법: 설문조사

2. 분석 결과: 설문조사 결과 날씨와 야외활동의 영향 정도를 보았을 때, 성별의 경우 차이가 없으나, 연령별, 학력별, 소득별로는 집단 간 조금씩 차이가 나타났다. 

3. TCI와 차이점: 

1) TCI와 가중치에서 차이점이 있다. 연중 KTCI = 2(2.59Cd + 2.09Ca + 3.22P + 1.16W + 0.95S) 

2) 계절별로 가중치 차이가 있는지에 대한 차이 검증결과 5가지 변수 모두에서 계절별 차이가 있는 것으로 나타났다.

3) 분석 결과에 따르면, 최고기온은 여름철, 평균기온은 겨울철, 강수량은 가을철, 풍속은 겨울철, 구름의 양은 봄철에 가장 가중치가 높은 것으로 나타났다.

의의: 계절별 기후 요소별 가중치의 차이가 존재한다는 것은 관광기후지수를 개발함에 있어서 한국의 경우, 계절을 고려한 계절별 관광기후지수 개발이 필요함을 보여주는 것을 의미한다.

4. 야외여가활동별 기후요소 가중치

계절에 따라 야외여가활동에는 차이가 있음.

잔차분석으로 수정된 잔차분석을 시도하였음. (절대값으로 2이상이면 특징적으로 간주할 수 있다. 출처:(노형진, 2004)) 

잔차분석 결과, 봄에는 자전거타기가 가장 많고 걷기가 적으며, 여름에는 피크닉이 많고, 걷기가 적으며, 가을에서 피크닉이 가장 많고 기타 활동이 적으며, 겨울에는 모두 가장 많은 활동비율을 차지하고 있으며, 겨울에는 걷기가 가장 많고 피크닉이 가장 적은 것으로 나타났다.

(후보: 휴식, 피크닉, 걷기(산책, 관람), 자전거 타기, 기타)

4. 계절별로 야외여가활동 간 기후요소 가중치의 차이가 존재하는 지 검토하였고, Scheffe 방법을 이용하여 사후검증 실시

결과: 봄에는 최고기온과 풍속, 여름에는 구름의 양, 겨울에는 풍속과 구름이 양이 여가활동 간에 통계적으로 유의미한 차이가 있는 것으로 나타났고, 가을에서 차이가 없는 것으로 나타났다.

5. 연중으로 보면 평균기온, 풍속, 구름의 양의 가중치가 활동에 따라 차이가 존재하는 것으로 나타났다.

6. 기후요소는 야외여가활동별로 각각 다른 가중치로 어느 정도 영향을 미치고 있음을 의미한다고 하겠다.

결론
--

본 연구결과는 도시공원과 같은 공간에서 이루어지는 여가활동에 대한 기후변화
예측을 통해 여가활동에 대한 예측 등에 사용될 수 있을 것으로 기대된다. 나아가
서는 본 연구에서 도출된 TCI는 우리나라의 관광자와 관광사업체에게 관광 및 여
가활동을 하기에 좋은(또는 나쁜) 시기와 공간적 분포를 분석하는데 기본적인 지표
로 활용될 수 있다. 특히 이 지표가 미래기후시나리오와 결합하여 사용하면 미래 
관광환경의 강점과 취약성과 같은 변화를 예측하는데 유용하게 사용될 수 있을 것
