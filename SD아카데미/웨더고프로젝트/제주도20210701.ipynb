```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

jeju_tour=pd.read_excel('제주.xls')
jeju1112=pd.read_csv('제주지방기상청(11-12).csv', parse_dates=['일시'])
jeju1314=pd.read_csv('제주지방기상청(13-14).csv', parse_dates=['일시'])
jeju1516=pd.read_csv('제주지방기상청(15-16).csv', parse_dates=['일시'])
jeju1718=pd.read_csv('제주지방기상청(17-18).csv', parse_dates=['일시'])
jeju1920=pd.read_csv('제주지방기상청(19-20).csv', parse_dates=['일시'])
```

```python
jeju_weather=pd.concat([jeju1112, jeju1314, jeju1516,  jeju1718, jeju1920], axis=0)
```

```python
jeju_weather.drop(['일기현상'], axis=1, inplace=True)
```

```python
jeju_weather=jeju_weather.fillna(0)
```

```python
jeju_weather['time'] = jeju_weather['일시'].dt.strftime('%Y%m%d')
jeju_weather['year_month'] = jeju_weather['일시'].dt.strftime('%Y%m')
jeju_weather['month_day'] = jeju_weather['일시'].dt.strftime('%m%d')
jeju_weather['year'] = jeju_weather['일시'].dt.strftime('%Y')
jeju_weather['month']=jeju_weather['일시'].dt.strftime('%m')

jeju_weather['time']=jeju_weather['time'].astype(int) 
jeju_weather['year_month']=jeju_weather['year_month'].astype(int)
jeju_weather['month_day']=jeju_weather['month_day'].astype(int) 
jeju_weather['year']=jeju_weather['year'].astype(int) 
jeju_weather['month']=jeju_weather['month'].astype(int)
```
