```python
def data_load_sep_local(path, name):
  #데이터 불러오고 내국인만 추출한 뒤, 컬럼 바꿔주기
  #path엔 "주요관광지점입장객데이터 경로 및 파일명, 확장자명" 선언
  #name : 해당 데이터를 저장할 변수명(예. "GangWon")

  #데이터 불러오기  
  locals()[f"{name}"] = pd.read_excel(path)
  #내국인만 추출한 뒤 저장
  locals()[f"{name}_local"] = locals()[f"{name}"][locals()[f"{name}"].iloc[:,3]=="내국인"]
  
  #년월이 "Unnamed:n" 으로 저장되어서 그냥 한번에 바꿈
  #저장될 필요가 없는 변수는 locals로 불러오지 않음
  dataCol = [col for col in locals()[f"{name}_local"].columns]
  dataNCol = [ncol for ncol in locals()[f"{name}"].iloc[0,:]]
  dataFCol = {}
  for og, nw in zip(dataCol, dataNCol):
    if ":" in og:
      dataFCol[og]=nw
    else:
      dataFCol[og]=og
  locals()[f"{name}_local"].rename(columns=dataFCol, inplace=True)
  
  #내국인 추출로 난장난 index 초기화
  locals()[f"{name}_local"].reset_index(drop=True, inplace=True)
  
  return locals()[f"{name}"], locals()[f"{name}_local"]

def sep_sgg(local_df, initial):
  #시군구로 지역범위가 넓은 곳의 데이터 분류
  #local_df : 위의 함수로 불러온 내국인만 있는 데이터
  #initial : 시군구별 분류한 데이터를 저장할 이니셜(예. "GW_춘천시" 중 "GW")
  
  globals()[f"{initial}_SGG"] = [sgg for sgg in local_df.iloc[:,1].unique()]

  for sgg in globals()[f"{initial}_SGG"]:
    globals()[f"{initial}_{sgg}"] = local_df[local_df.iloc[:,1]==sgg]
    globals()[f"{initial}_{sgg}"].reset_index(drop=True, inplace=True)
    print(f"'{initial}_{sgg}'의 shape은 ", globals()[f"{initial}_{sgg}"].shape, "입니다.")
  #globals를 활용해 앞에 변수 지정 없이도 불러올 수 있게 설정함
```

```python
def Y_YM_Visitor(Ydf, YMdf, Y_duration):
  # Ydf : 연별 데이터, YMdf : 월별 데이터
  # Y_duration : 종료일 기준 확인 범위(단위 : 년)
  # 연별 데이터로는 해당되지 않는 관광지도 월별 데이터로 확인하면 결측치가 있는 경우가 있음

  check_Ydf = Ydf.iloc[-Y_duration:, :].copy()
  check_YMdf = YMdf.iloc[-(Y_duration*12):, :].copy()

  del_Ycol_idx = []   #연별 데이터 중 결측년이 존재하는 관광지
  del_YMcol_idx = []  #월별 데이터 중 결측월이 존재하는 관광지

  for check in range(check_Ydf.shape[1]):
    for TorF, TorF2 in zip(check_Ydf.iloc[:, check].isna(), check_Ydf.iloc[:, check].isin([0])):
      if TorF==True:
        del_Ycol_idx.append(check)
      if TorF2==True: #NaN이 아닌 int(0)으로 입력된 결측치도 있었음
        del_Ycol_idx.append(check)
      
    for TorF, TorF2 in zip(check_YMdf.iloc[:, check].isna(), check_YMdf.iloc[:, check].isin([0])):
      if TorF==True:
        del_YMcol_idx.append(check)
      if TorF2==True: #NaN이 아닌 int(0)으로 입력된 결측치도 있었음
        del_YMcol_idx.append(check)

  del_Ycol_idx = sorted(list(set(del_Ycol_idx)))    #3개년 true값 중복으로 들어갔을 check 넘버 제거
  del_YMcol_idx = sorted(list(set(del_YMcol_idx)))  #36개월 true값 중복으로 들어갔을 check 넘버 제거

  print(f"{Y_duration}개년 중 결측년 존재하는 관광지 수 : {len(del_Ycol_idx)}")
  print(f"{Y_duration*12}개월 중 결측월 존재하는 관광지 수 : {len(del_YMcol_idx)}")

  clean_Ycol_idx = []   #연별 데이터 중 활용 가능한 관광지
  clean_YMcol_idx = []  #월별 데이터 중 활용 가능한 관광지

  for all in range(check_Ydf.shape[1]):
    # 연별 데이터가 있는 경우와 연월별 데이터가 있는 경우가 이어지는 가정이 아니기 때문에, if~elif~가 아닌 if~ if~로 조건반복문 작성
    if all not in del_Ycol_idx:
      clean_Ycol_idx.append(all)

    if all not in del_YMcol_idx:
      clean_YMcol_idx.append(all)

  YR_Clean = check_Ydf.iloc[:, clean_Ycol_idx]
  YM_Clean = check_YMdf.iloc[:, clean_YMcol_idx]

  print(f"{Y_duration}개년 중 활용 가능한 관광지 수 : {YR_Clean.shape[1]}")
  print(f"{Y_duration*12}개월 중 활용 가능한 관광지 수 : {YM_Clean.shape[1]}")

  YR_Clean.reset_index(inplace=True)
  YR_Clean.rename(columns={"index":"일시"}, inplace=True)
  YM_Clean.reset_index(inplace=True)
  YM_Clean.rename(columns={"index":"일시"}, inplace=True)

  return YR_Clean, YM_Clean
```

현재 기상데이터는 2957에 대한 정보만 존재하므로, 결측일(혹은 결측월)에 대한 정보를 알고 있어야 해석의 용의성 및 모델의 정확도가 올라감

윤년이란 상황을 가정하고, 매월 달라지는 2월의 일자까지 확인하기 위해서 우선 datetime 모듈을 활용해서 일 계산 및 결측치 확인에 사용

```python
def Date_Duration():
  print(f"{'='*10}{'<'*5}\tInitializing\t{'>'*5}{'='*300}")
  sY, sM, sD  = map(int, input("시작일 : ").split(" "))
  startDate = datetime(sY, sM, sD)

  eY, eM, eD  = map(int, input("종료일 : ").split(" "))
  endDate = datetime(eY, eM, eD)

  duration = (endDate-startDate).days
  print(f"조사 기간의 기상 데이터의 일 수는 총 {duration}일이어야 합니다.\n")
  #결측일이 없으려면, duration 값과 기상 데이터의 rows가 동일해야 함

  return startDate, endDate, duration


def Missing_Date_Check(dt, datecol, duration, startDate=None, check_type="%Y%m%d"):
  print(f"{'='*10}{'<'*5}\tmissing dates\t{'>'*5}{'='*300}")
  missing_days = []
  missing_day = 0

  for day in range(duration):
    check_date = startDate+timedelta(days=day)
    if dt[datecol][day-missing_day].strftime(check_type)==check_date.strftime(check_type):
      continue
    elif dt[datecol][day-missing_day].strftime(check_type)!=check_date.strftime(check_type):
      missing_day+=1
      missing_days.append(check_date) # 활용을 쉽게 활용할 수 있도록 datetime으로 추가

  if len(missing_days)!=0:
    print(f"*경고*\n결측일이 발견되었습니다.")
    print(f"{[mdate.strftime(check_type) for mdate in missing_days]}의 결측일 여부를 확인하세요.\n")

  return missing_days


def Re_val_MDate(dt, datecol, missing_days, check_type="%Y%m%d"):
  print(f"{'='*10}{'<'*5}\tre-validation\t{'>'*5}{'='*300}")
  cnt = 0
  re_val_date = []

  for idx, mdate in enumerate(missing_days):
    if len(dt[dt[datecol]==mdate.strftime(check_type)].index)!=0:
      re_val_date.append(idx)
    elif len(dt[dt[datecol]==mdate.strftime(check_type)].index)==0:
      cnt += 1

  if cnt==len(missing_days):
    print("재검증 결과 : 상동")
    print(f"해당 기상 데이터에 존재하는 결측일은 총 {cnt}일입니다.\n")
  elif cnt!=len(missing_days):
    for pop_idx in re_val_date:
      missing_days.pop(pop_idx)
    print("재검증 결과 : 상이")
    print(f"해당 기상 데이터에 존재하는 결측일은 총 {len(missing_days)}일입니다.\n")

  return missing_days


def Cnts_MDays_in_YearMonth(missing_days, check_ym_type="%Y%m", return_ym_type="%Y년%m월"):
  print(f"{'='*10}{'<'*5}\tCounts of Missing days by Months\t{'>'*5}{'='*300}")

  if check_ym_type!=return_ym_type:
    missing_days_by_ym = [mdates.strftime(return_ym_type) for mdates in missing_days]
    missing_ym = sorted(list(set(missing_days_by_ym)))
    for mYM in missing_ym:
      globals()[f"missing_{mYM}"] = missing_days_by_ym.count(mYM)
      print(f'{mYM}의 결측일 수 : {globals()[f"missing_{mYM}"]}\n')

  else :
    missing_days_by_ym = [mdates.strftime(check_ym_type) for mdates in missing_days]
    missing_ym = sorted(list(set(missing_days_by_ym)))
    for mYM in missing_ym:
      globals()[f"missing_{mYM}"] = missing_days_by_ym.count(mYM)
      print(f'{mYM}의 결측일 수 : {globals()[f"missing_{mYM}"]}\n')

  return missing_ym
```

```python
#when dtype is int or float
def zero_Check(dt, col):
  zrcnt = dt[col][df[col]==0].count()
  zridx = dt[df[col]==0].index
  print(f'zero counts of {col} : {zrcnt}\n')
  print(f'\t{zridx}')
  return zrcnt, zridx

#when dtype is object
def null_Check(dt, col):
  zrcnt = 0
  zridx = []
  for idx, cont in enumerate(dt[col]):
    if type(cont)==float:
      zrcnt += 1
      zridx.append(idx)
  print(f'null values in {col} : {zrcnt}\n')
  print(f'\t{zridx}')
  return zrcnt, zridx

#no matter dtype numeric nor category, check value is string typed "N/A"
def NONE_Check(dt, col):
  nacnt = dt[col][df[col]==("None" or "N" or "NONE" or "N/A")].count()
  naidx = dt[df[col]==("None" or "N" or "NONE" or "N/A")].index
  print(f'"N/A" counts of {col} : {nacnt}\n')
  print(f'\t{naidx}')
  return nacnt, naidx
```

```python
#정열님 sns 함수
def graph(data = None, x = None, y = None, hue = None, option = None):
  data = data[data[y] != np.nan]
  data[y] = data[y].astype('float64')
  if option == 'line':
    plt.figure(figsize = (16,8))
    sns.lineplot(data = data, x = x, y = y, hue = hue)
  elif option == 'bar':
    plt.figure(figsize = (16,8))
    sns.barplot(data = data, x = x, y = y, hue = hue)
  elif option == 'box':
    plt.figure(figsize = (16,8))
    sns.boxenplot(data = data, x = x, y = y, hue = hue)
  elif option == 'dist':
    plt.figure(figsize = (16,8))
    sns.distplot(data = data, x = x, y = y, hue = hue)
  elif option == 'count':
    plt.figure(figsize = (16,8))
    sns.countplot(data = data, x = x, y= y, hue = hue)
  else:
    pass
  plt.title(f'{y}')
  plt.show()
```
