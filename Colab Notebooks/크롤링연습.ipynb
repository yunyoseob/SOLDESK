```python
import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook
import pandas as pd

df=pd.DataFrame(columns=["이름","내용"])

def enter_url(df, name, url): 
    response=requests.get(url)
    html=response.text
    soup=BeautifulSoup(html, "html.parser")
    df=df.append({"이름":name, "내용":soup}, ignore_index=True)
    return df

df2=enter_url(df, "카르도", "https://www.wanted.co.kr/wd/90057")
```
