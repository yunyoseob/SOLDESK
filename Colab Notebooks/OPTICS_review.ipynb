군집 분석(cluster analysis)은 각 개체에 대해 관측된 여러 개의 변수 $(x_{1}, x_{2}, ..., x_{p}) $ 값들로 부터  $n$개의 개체를 유사한 성격을 가지는 몇 개의 군집으로 집단화 하고, 형성된 군집들의 특성을 파악하여 군집들 사이의 관계를 분석하는 다변량 분석기법이다

병합 군집(Aggolmerative Clustering)은 각각의 데이터 포인트를 하나의 클러스터로 지정하고, 지정된 개수의 클러스터가 남을 때까지 가장 비슷한 두 클러스터를 합쳐 나가는 알고리즘이다.

병합 군집은 계층적 군집을 만든다.

계층적 군집(hierarchical clustering)은 가장 유사한 개체를 묶어 나가는 과정을 반복하여 원하는 개수의 군집을 형성하는 방법이다.

계층적 군집은 두 개체(또는 군집) 간의 거리(또는 비유사성)에 기반하여 군집을 형성해 나가므로 거리에 대한 정의가 필요하다.

DBSCAN은 각각의 데이터들에 대해 이웃한 데이터와의 밀도를 계산하면서 불특정한 모양의 클러스터를 생성한다. 

OPTICS clustering은 DBSCAN처럼 밀도를 기반으로 군집을 형성하는 방법이다.

**What is Density-based clustering?**

Density-based clustering methods detect groups of similar observations based on their distance to a given number of their nearest neighbors. 

**K-means vs Density-based clustering**

In constrast with other clustering methods a s K-means, they do not expect the observed data to follow Gaussian or other parametric distributions and they can thus detect groups of various shapes.

- 출처: opticskxi: OPTICS K-XI Density-Based Clustering (Thomas Charlon)

- https://cran.r-project.org/web/packages/opticskxi/vignettes/opticskxi.pdf





OPTICS clustering algorithm이란 밀도기반 클러스팅중 하나로, 민코우스키 거리를 사용하는 알고리즘이다.

- 민코우스키(Minkowski) 거리

$$ d(i,j) = [\sum_{f=1}^{p} (x_{if}-x_{jf})^m]^{1/m} $$

- 출처: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.OPTICS.html?highlight=cluster#sklearn.cluster.OPTICS


**참고**

DBSCAN: by default the euclidean distance is used

- 유클리드(Euclidian) 거리

$$ d(i,j) = \sqrt{\sum_{f=1}^{p} (x_{if}-x_{jf})^2} $$

**OPTICS**

OPTICS is density-based algorithm that produces an ordering and a distance profile of observations, similar to tree-like den-drogram and enables the detection of clusters of varying densities with the cluster extraction methods OPTICS Xi.

- 출처: opticskxi: OPTICS K-XI Density-Based Clustering (Thomas Charlon)

- https://cran.r-project.org/web/packages/opticskxi/vignettes/opticskxi.pdf

DBSCAN vs OPTICSKXI
--

Ordering points to identify the clustering structure (OPTICS) is an algorithm for finding density-based clusters in spatial data. It was presented by Mihael Ankerst, Markus M. Breunig, Hans-Peter Kriegel and Jörg Sander. Its basic idea is similar to DBSCAN, but it addresses one of DBSCAN's major weaknesses: the problem of detecting meaningful clusters in data of varying density. 

출처: https://en.wikipedia.org/wiki/OPTICS_algorithm


DBSCAN
--

```python
install.packages('dbscan')
```

```python
library(dbscan)
```

```python
dbscan_shapes <- dbscan::dbscan(iris[1:2], eps=0.15)
gg_shapes <- cbind(iris[1:2], Clusters=dbscan_shapes$cluster)%>%ggpairs(group="Clusters")
```

```python
data('DS3', package = 'dbscan')
```

```python
dbscan_ds3<-dbscan::dbscan(DS3, minPts = 25, eps=12)
gg_ds3<-cbind(DS3, Clusters=dbscan_ds3$cluster)%>%ggpairs(group='Clusters')
```

```python
install.packages('cowplot')
```

```python
library(cowplot)
```

```python
cowplot::plot_grid(gg_shapes, gg_ds3, nrow=2, labels=c('(a)','(b)'), label_x=0.9)
```

The basic approach of OPTICS is similar to DBSCAN, but instead of maintaining known, but so far unprocessed cluster members in a set, they are maintained in a priority queue.

- priority queue: In computer science, a priority queue is an abstract data type similar to a regular queue or stack data structure in which each element additionally has a "priority" associated with it.

**DBSCAN vs OPTICS**

1. OPTICS (Ordering Points To Identify the Clustering Structure), closely related to DBSCAN, finds core sample of high density and expands clusters from them. Unlike DBSCAN, keeps cluster hierarchy for a variable neighborhood radius. Better suited for usage on large datasets than the current sklearn implementation of DBSCAN.

2. DBSCAN uses a fixed distance threshold and thus can not detect clusters of varying
densities. 

3. Finds core samples of high density and expands clusters from them. This example uses data that is generated so that the clusters have different densities. The OPTICS is first used with its Xi cluster detection method, and then setting specific thresholds on the reachability, which corresponds to DBSCAN. We can see that the different clusters of OPTICS’s Xi method can be recovered with different choices of thresholds in DBSCAN.

OPTICS는 밀도 기반의 클러스터를 탐색하기 위한 알고리즘으로, DBSCAN과 유사하지만, 다양한 밀도를 가진 데이터에서 클러스터를 잘 찾아내지 못하는 문제를 해결하기 위해 고안 되었다. 데이터들을 정렬하여 가까운 point들이 이웃이 될 수 있도록 하며, 각 point가 더 밀집된 cluster에 포함될 수 있도록 한다.

**Why Tree??**

Hierarchical algorithms decompose a database D of objects into several levels of nested partitionings (clusterings), represented by a dendrogram

- dendrogram(tree)는 계층적 이진 군집 트리에 대한 덴드로그램 플롯을 생성하며, 계층적 트리의 데이터 점을 연결하는 다수의 U자형 선으로 구성됩니다.
