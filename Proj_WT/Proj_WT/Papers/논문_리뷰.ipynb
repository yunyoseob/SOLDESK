## 기상요인이 관광에 미치는 영향(제주도)
http://www.riss.kr/search/detail/DetailView.do?p_mat_type=be54d9b8bc7cdb09&control_no=3e58a59f19d774f3ffe0bdc3ef48d419&outLink=K

- 각 항목들에 대해서 ANOVA(일원변량분석) 실시 
- Seheffe의 다중 범위 검정을 통하여 유의확률 0.05 에서 유의한 차이를 검정
- 계절별로 분석


각 공항별(제주, 김포, 김해) 항공기 결항률과 결항 요인들 조사
- $결향률 = \frac{요인별결항횟}{수운항횟수} \times 100  $

결항 요인중 기상요인이 차지하는 비율

제주도 소재 관광지 두 곳을 선정하여 방문객수와 기상정보의 상관관계, 그리고 기상 요인이 관광지 선택에 어떠한 영향을 어느 정도 미치는지 조사

기상요소(최저기온, 최고기온, 평균기온, 강수량, 운량, 습도, 풍속)

실외형 관광지와 실내형 관광지 동일조건에서 비교함

기상요인과 기상현상으로 인한 불쾌한 경험, 제주도 방문횟수, 관광지 방문횟수에 따른 관광지 선택의 결정요인의 차이
- t-test 와 ANOVA
- Duncan의 다중범위 검정



## 날씨 및 요일 특성이 관광정보시스템 이용에 미치는 영향

기상상태나 요일특성이 원인이 되어 발생하는 일반적인 수요이론에 반하는 이상현상이 다양한 분야에 존재 
--> 날씨 및 요일 특성이 관광정보시스템에도 영향이 있을까?

계절, 휴일 같은 날짜 요인들이 관광정보시스템의 접속자수에 미치는 영향

강수, 강설, 운량 같은 날씨요인이 관광정보시스템의 접속자수에 미치는 영향

- 날씨요인 : 강수, 적설, 운량
- 요일요인 : 계절, 휴일 여부

t-test, ANOVA 실시 유의수준 0.05






## 해외방문관광객의 목적지 선택에 대한 기후의 영향

PTCM 분석법 사용
- 하나의 목적지에 대한 경제적 총가치를 산출하기 위하여 조사하는 방법
- 다수의 목적지를 조사하는 방법으로 목적지 특성이 변화 할 때 가치 변화를 측정하는 방법

- 기후 변화의 영향을 조사하는 경우에는 두 번째 모델이 유용

월별 기후 자료로 분석

종속변수 : 우리나라 해외 관광객수(상용 여행, 유학등과 같은 이유로 해외를 방문하는 관광객은 제외)

여행비용 : 특정 여행사의 국적기 이용 패키지별 비용과 저가항공 이용 패키지별 비요을 평균하여 산출

대리변수 : 국가별 수도간의 직선거리 이용

미국 달러로 조사된 국가별 1인당 GDP

국가별 인구수와 인구밀도

국가 안정도 지수

국가별 기후 자료

각 국가의 월평균 기후 자료(월평균최고기온, 월평균최저기온, 월강수량, 월평균강수일수)
