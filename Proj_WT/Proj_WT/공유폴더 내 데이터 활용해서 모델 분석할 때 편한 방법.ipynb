### 정열님께서 정리해준 xlsx 파일로 기상청 관리관서와 지점 불러오기

```python
%cd "{SV2}"
!ls
```

```python
data = pd.read_excel(f'{SV2}/branch.xlsx')
data['지점'] = data['지점'].astype('str')
office = data['관리관서'].unique()
print(office)
branch = [] # branch : 기상청 지점
for j in office: #office : 기상청 관리관서
  KMA = {}
  for i in data.index:
    if data.loc[i,'관리관서'] == j:
      KMA[data.loc[i,'지점명']] = data.loc[i,'지점']
  branch.append(KMA)
```

* 종관기상관측 장비로 관측한 일 기상자료를 조회하는 서비스

#### 기상청 관리관서 및 지점 조회 코드

```python
for of, br in zip(office, branch):
  print(of, br)
  for name, code in br.items():
    print(name, code)
```
