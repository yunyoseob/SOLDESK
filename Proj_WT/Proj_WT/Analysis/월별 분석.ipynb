```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm

x_column = 충북데이터.columns[:]
y_column = final_data.columns[len(충북데이터.columns):]
print(x_column, y_column)
```

```python
# 지금 우리가 있는 데이터인 2017~2019의 데이터를 가지고 분석을 진행해보기로 하자.
# train과 test로 데이터를 나눈다.
train = final_data[(final_data['date'] >= '201701') & (final_data['date'] <= '201812')]
test = final_data[(final_data['date'] >= '201901')]

# 분석하기 용이하게 데이터를 학습 데이터와 타겟 데이터로 나눠준다.
X_train = train[x_column]
y_train = train[y_column]

X_test = test[x_column]
y_test = test[y_column]
```

```python
# sns로 그릴 수 있는 그래프를 만들어보자.
def graph(data = None, x = None, y = None, hue = None, option = None):
  data = data[(data[y] != np.nan) | (data[y] > 0)]
  data[y] = data[y].astype('float64')
  if option == 'line':
    plt.figure(figsize = (16,8))
    sns.lineplot(data = data, x = x, y = y, hue = hue)
  elif option == 'bar':
    plt.figure(figsize = (16,8))
    sns.barplot(data = data, x = x, y = y, hue = hue)
  elif option == 'box':
    plt.figure(figsize = (16,8))
    sns.boxenplot(data = data, x = x, y = y, hue = hue)
  elif option == 'dist':
    plt.figure(figsize = (16,8))
    sns.distplot(data = data, x = x, y = y, hue = hue)
  elif option == 'count':
    plt.figure(figsize = (16,8))
    sns.countplot(data = data, x = x, y= y, hue = hue)
  else:
    pass
  plt.xticks(rotation=90)
  plt.title(f'{y}')
  plt.show()
```
