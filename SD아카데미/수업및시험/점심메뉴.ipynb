```python
import numpy as np
import random

class today_lunch:
    def choice(a,b):
        x=[]
        x.append(a)
        x.append(b)
        return x

    def menu(a,b):
        x=today_lunch.choice(a,b)
        y=random.choice(x)
        return y
```
