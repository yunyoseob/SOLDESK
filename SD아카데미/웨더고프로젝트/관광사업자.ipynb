LocalData 관광 (유람선, 테마파크, 온천, 놀이공원 등 관광객에게 시설이용을 제공하거나 서비스하는 업소 정보)
--

https://www.localdata.go.kr/devcenter/dataDown.do?menuNo=20001

1) 관광궤도업: 케이블카 등 주변 관람과 운송에 적합한 시설을 갖추어 제공하는 업소정보

2) 관광사업자: 관광숙박업, 국제회의업, 일반여행업 등 관광사업을 경영하는 업소정보

3) 관광유람선업: 유람선 크루즈와 같은 선박을 이용하여 관광객에게 관광을 제공하는 업소정보

4) 국제회의시설업: 컨벤션센터 등 대규모 관광 수요를 유발하는 국제회의를 개최할 수 있는 시설을 설치하여 운영하는 업소정보

5) 박물관, 미술관: 고고학적자료, 역사적 유물, 예술품 또는 미술품 등 자료를 수집.보존.진열하여 일반인에게 전시하는 시설정보

6) 시내순환관광업: 버스를 이용하여 관광객에게 시내와 그 주변 관광지를 정기적으로 순회하면서 관광을 제공하는 업소정보

7**) 유원시설업(기타): 키즈카페 등 안전성 검사 대상이 아닌 유기시설 또는 유기기구를 설치하여 운영하는 업소정보

8) 일반유원시설업: 테마파크 등 안전성 검사대상의 놀이기구를 한 종류 이상 설치하여 운영하는 업소정보

9) 전문휴양업: 온천 등 관광객의 휴양이나 여가 선용을 위하여 숙박시설이나 음식점 시설을 갖추어 제공하는 업소 정보

10**) 전통사찰: 불교 신앙의 대상으로서의 형상을 봉안하고 승려가 수행하며, 신도를 교화하기 위한 시설 및 공간 정보

11) 종합유원시설업: 놀이공원 등 안전성 검사대상의 놀이기구를 6종류 이상 설치하여 운영하는 업소 정보

12) 종합휴양업: 리조트 등 숙박시설 또는 음식점시설을 갖추고 전문휴양시설 등 두 종류 이상의 시설을 갖추거나, 한 종류 이상의 시설

13) 지방문화원: 지역문화의 진흥을 위한 지역문화사업을 수행하기 위하여 설립된 법인 정보
